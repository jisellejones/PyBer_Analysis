```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = "Boston")

# Create labels for the x- and y-axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Set the x limit
plt.xlim(0, 12)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend.
plt.legend()

# Change figure size to 20 by 10
plt.figure(figsize = (20, 10))
```

```python
#Object oriented approach

# Create the plot
fig, ax = plt.subplots(figsize = (20, 10))
ax.plot(x_axis, y_axis, color = "green", marker = "d", linewidth = 2, label = 'Boston')

# Add y- and x-axis labels.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Add a y-axis limit
ax.set_ylim(0, 50)

# Add grid lines.
ax.grid()

# Add a title
ax.set_title("PyBer Fare by Month")
```

```python
# BAR GRAPH
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color = "green", label = "Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create a title.
plt.title("PyBer Fare by Month")

#Add the legend.
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis)
```

```python
# Make axes opposites - this is a bad grid
plt.barh(y_axis, x_axis)
```

```python
# Invert axes to have months in the right order (on first horizontal grid)
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.gca().invert_yaxis()

#Show legend
plt.legend()

#Add Title
plt.title("PyBer Fare by Month")

# Add x- and y- axes labels
plt.xlabel("Fare ($)")
plt.ylabel("Date")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "k", label = "New York")

# Add legend
ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "k", label = "New York")

# Add legend
ax.legend()
```

```python
# To switch axes in object-oriented graph, switch positions
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```
